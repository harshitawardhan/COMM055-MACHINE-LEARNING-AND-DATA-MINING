```python
#next we will remove outliers using the Z score method
#https://www.analyticsvidhya.com/blog/2021/05/feature-engineering

def handle_and_plot_outliers(data, outlier_columns):
    updated_data = data.copy()

    fig, axes = plt.subplots(len(outlier_columns), 2, figsize=(10, 5 * len(outlier_columns)))
    fig.suptitle('Distribution Before and After Outlier Handling Using Z-scores', fontsize=16)

    for idx, column in enumerate(outlier_columns):
        original_data = updated_data[column].copy()

        mean = original_data.mean()
        std = original_data.std()

        highest_allowed = mean + 3 * std
        lowest_allowed = mean - 3 * std

        updated_data[column] = np.where(
            original_data > highest_allowed,
            highest_allowed,
            np.where(
                original_data < lowest_allowed,
                lowest_allowed,
                original_data
            )
        )

        sns.histplot(original_data, bins=30, kde=True, ax=axes[idx, 0], color='blue')
        axes[idx, 0].set_title(f'Original {column}')
        axes[idx, 0].set_xlabel(column)
        axes[idx, 0].set_ylabel('Frequency')

        sns.histplot(updated_data[column], bins=30, kde=True, ax=axes[idx, 1], color='green')
        axes[idx, 1].set_title(f'Adjusted {column}')
        axes[idx, 1].set_xlabel(column)
        axes[idx, 1].set_ylabel('Frequency')

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

    return updated_data

outlier_columns = ['age', 'fnlwgt', 'capital-gain', 'capital-loss', 'hours-per-week']
Dataset = handle_and_plot_outliers(Dataset, outlier_columns)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

def impute_missing_values(df, target_columns):
    df = df.copy()
    numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()
    
    for col in target_columns:
        if df[col].isna().any():
            train_data = df[df[col].notna()]
            test_data = df[df[col].isna()]

            if train_data.empty or test_data.empty:
                continue

            X_train = train_data[numerical_columns]
            y_train = train_data[col]
            X_test = test_data[numerical_columns]

            model = RandomForestClassifier(n_estimators=100, random_state=42)
            model.fit(X_train, y_train)
            
            predicted_values = model.predict(X_test)
            df.loc[df[col].isna(), col] = predicted_values
            print(f"Missing values imputed for column: {col}")

    return df
target_columns = ['workclass', 'occupation', 'native-country']
second_encoded_dataset = impute_missing_values(second_encoded_dataset, target_columns)
```

```python
def plot_distributions(df_before, df_after):
    numerical_data_before = df_before.select_dtypes(include=[np.number])
    numerical_data_after = df_after.select_dtypes(include=[np.number])
    
    num_vars = numerical_data_before.columns.size
    num_cols = 2
    num_rows = int(np.ceil(num_vars / num_cols))
    
    fig, axes = plt.subplots(num_rows * 2, num_cols, figsize=(15, 6 * num_rows), squeeze=False)
    axes = axes.flatten()
    
    for i, column in enumerate(numerical_data_before.columns):
        ax_before = axes[i * 2]
        sns.histplot(numerical_data_before[column], kde=True, color='skyblue', bins=30, ax=ax_before)
        mean_value_before = numerical_data_before[column].mean()
        ax_before.axvline(mean_value_before, color='red', linestyle='--', label=f'Mean: {mean_value_before:.2f}')
        ax_before.set_title(f'Before Downsampling: {column}')
        ax_before.set_xlabel(column)
        ax_before.set_ylabel('Frequency')
        ax_before.legend()
        
        ax_after = axes[i * 2 + 1]
        sns.histplot(numerical_data_after[column], kde=True, color='lightgreen', bins=30, ax=ax_after)
        mean_value_after = numerical_data_after[column].mean()
        ax_after.axvline(mean_value_after, color='red', linestyle='--', label=f'Mean: {mean_value_after:.2f}')
        ax_after.set_title(f'After Downsampling: {column}')
        ax_after.set_xlabel(column)
        ax_after.set_ylabel('Frequency')
        ax_after.legend()
    
    for ax in axes[num_vars * 2:]:
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()

# Plot distributions before and after downsampling
plot_distributions(df_majority, df_majority_downsampled)
```
